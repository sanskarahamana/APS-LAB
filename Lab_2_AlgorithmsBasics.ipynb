```python
# Exercise 1
def func1(a, b):
  # This function calculates the product of two numbers (a * b)
  # using repeated addition. It adds 'a' to 'result' 'b' times.
  # Time Complexity: O(b) - the loop runs 'b' times.
  result = 0
  for i in range(0, b):
    result += a
  return result
```

```python
# Sample function call: func1(4, 3) should return 12
x = print(func1(4, 3))
# The function performs multiplication using repeated addition. Its time complexity is O(b).
```

```python
# Exercise 2
def func2(a, b):
  result = 1
  while b > 0:
    result = result * a
    b = b - 1
  return result
```

```python
# Exercise 4
def func4(a, b):
    b.sort() # Time complexity for this function is O(Nlog N) where N = len(b)
    intersect = 0
    for x in a:
        if binary_search(b, x) >= 0:
            intersect = intersect + 1
    return intersect

def binary_search(arr, target):
    low, high = 0, len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1
```

```python
# Sample function call: func4([1, 2, 3, 4], [2, 4, 6, 8])
```

```python
# Exercise 5
def func5(n):
    guess = 1
    while guess * guess <= n:
        if guess * guess == n:
            return guess
        guess = guess + 1
    return None
```

```python
# Exercise 6
def func8(A):
    func8_helper(A, 0, len(A) -1)

def func8_helper(A, i, j):
    #Input: An array A and two indices i and j.
    n = j - 1 + 1
    if n ==2:
        if A[i] > A[j]:
            # Swap A[i] and A[j]
            A[i], A[j] = A[j], A[i]
        elif n > 2:
            m = n // 3
            func8_helper(A, i, j - m)
            func8_helper(A, i + m, j)
            func8_helper(A, i, j - m)
        return A
```

```python
# Exercise 7
def func12(arr, k):
    arr_len = len(arr)
    k = k % arr_len
    # Initialize a new array of the same length as 'arr'
    custom_arr = [0] * arr_len

    for i in range(arr_len):
        mystery_index = (i + k) % arr_len
        custom_arr[mystery_index] = arr[i]
    return custom_arr

```

```python
# write a function named as sum_arr to find sum of elements in given array
def sum_arr(arr):
  total = 0
  for elements in arr:
    total += elements
    return total
```

```python
def max_min(arr):
  max_element = arr[0]
  min_element = arr[0]
  for elements in arr:
    if elements>max_element:
      max_element=elements
    elif elements<min_element:
        min_element=elements
  return max_element,min_element

arr = [1,23,45,67,89,12,33,44]
print(max_min(arr))
```

```python
# write a function named as Counting_EvenOdd to count number of even and odd elements in an array
def counting_evenodd(arr):
  even_count = 0
  odd_count = 0
  for elements in arr:
    if elements % 2 == 0:
       even_count +=1
    elif elements % 2 != 0:
       odd_count +=1
    return even_count,odd_count

arr = [12,1234,45,67,1]
print(counting_evenodd(arr))
```
