# Shortest Path Using Dijkstra Algorithm in Python
Problem Description:
Given a weighted graph with non negative edge weights. Find the shortest distance from a source vertex to all other vertices using Dijkstra algorithm

```python
def get_user_graph():
    print("Enter the graph as a dictionary string (e.g., A B:4,C:2;B A:4,C:5,D:10):")
    user_input = input()
    graph = {}
    try:
        nodes_str = user_input.split(';')
        for node_str in nodes_str:
            if not node_str.strip():
                continue
            parts = node_str.split(' ', 1)
            node = parts[0].strip()
            if len(parts) > 1:
                neighbors_str = parts[1].strip()
                neighbors = {}
                for neighbor_pair in neighbors_str.split(','):
                    if not neighbor_pair.strip():
                        continue
                    neighbor_parts = neighbor_pair.split(':')
                    if len(neighbor_parts) == 2:
                        neighbor_node = neighbor_parts[0].strip()
                        weight = int(neighbor_parts[1].strip())
                        neighbors[neighbor_node] = weight
                graph[node] = neighbors
            else:
                graph[node] = {}
    except Exception as e:
        print(f"Error parsing graph input: {e}")
        return None
    return graph

user_graph = get_user_graph()

if user_graph:
    print("User-defined graph:")
    for node, edges in user_graph.items():
        print(f"{node}: {edges}")

    # Get source node from user
    while True:
        start_node = input("Enter the starting node: ").strip().upper()
        if start_node in user_graph:
            break
        else:
            print(f"Node '{start_node}' not found in the graph. Please enter a valid starting node.")

    # Run Dijkstra's algorithm with the user-defined graph and start node
    distances, predecessors = dijkstra(user_graph, start_node)

    print(f"\nShortest distances and paths from node {start_node}:")
    for node, distance in distances.items():
        if distance == float('inf'):
            print(f"Path to {node}: No path (unreachable)")
        else:
            path = get_shortest_path(predecessors, start_node, node)
            print(f"Path to {node}: {' -> '.join(path)} (Distance: {distance})")
```

#Theory
Dijkstra algorithm finds shortest distance from one source node to all other nodes in a weighted graph with non negative weights. Dictionary structure stores nodes as keys and neighbor nodes with weights as values.

##Algorithm Steps

1. Represent graph using dictionary.
2. Set distance of every node as infinity.
3. Set distance of source node as 0.
4. Select node with smallest distance among unvisited nodes.
5. Update distance of neighbor nodes.
6. Repeat until all nodes get visited.

```python
def dijkstra(graph, start):
    distances = {}
    predecessors = {}
    visited = set()

    for node in graph:
        distances[node] = float('inf')
        predecessors[node] = None

    distances[start] = 0

    while len(visited) < len(graph):
        min_node = None
        min_distance = float('inf')

        for node in graph:
            if node not in visited and distances[node] < min_distance:
                min_distance = distances[node]
                min_node = node

        if min_node is None: # All remaining nodes are unreachable
            break

        visited.add(min_node)

        for neighbor, weight in graph[min_node].items():
            new_distance = distances[min_node] + weight

            if new_distance < distances[neighbor]:
                distances[neighbor] = new_distance
                predecessors[neighbor] = min_node

    return distances, predecessors

def get_shortest_path(predecessors, start, target):
    path = []
    current = target
    while current is not None:
        path.insert(0, current)
        current = predecessors[current]
    if path[0] == start: # Check if path starts from source
        return path
    return [] # No path found
```

```python
graph = {
'A': {'B': 4, 'C': 2},
'B': {'A': 4, 'C': 5, 'D': 10},
'C': {'A': 2, 'B': 5, 'D': 3},
'D': {'B': 10, 'C': 3}
}

result = dijkstra(graph, 'A')

for node in result:
    print("Distance from A to", node, "=", result[node])
```
