# Greedy Knapsack Problem (Fractional)
In this lab, we explore the Fractional Knapsack Problem, a fundamental optimization problem that demonstrates the Greedy Algorithmic Paradigm.
We will implement it using greedy selection (based on value-to-weight ratio) and analyze its time and space complexity.

```python
class Item:
    def __init__(self, value, weight):
        self.value = value
        self.weight = weight
        self.ratio = value / weight

    def __repr__(self):
        return f"Item(V={self.value}, W={self.weight}, Ratio={self.ratio:.2f})"

def fractional_knapsack(capacity, items):
    items.sort(key=lambda x: x.ratio, reverse=True)

    total_value = 0.0
    current_capacity = capacity
    knapsack_contents = []

    for item in items:
        if current_capacity <= 0:
            break

        if item.weight <= current_capacity:
            total_value += item.value
            current_capacity -= item.weight
            knapsack_contents.append(f"Took full {item.weight}kg of item with value {item.value} (Ratio: {item.ratio:.2f})")
        else:
            fraction = current_capacity / item.weight
            value_taken = item.value * fraction
            weight_taken = item.weight * fraction
            total_value += value_taken
            current_capacity = 0
            knapsack_contents.append(f"Took {weight_taken:.2f}kg ({fraction:.2f} fraction) of item with value {item.value} (Ratio: {item.ratio:.2f})")

    return total_value, knapsack_contents
```

```python
# Example Usage:
# Define items with their value and weight
items = [
    Item(60, 10),  # Item 1: Value=60, Weight=10
    Item(100, 20), # Item 2: Value=100, Weight=20
    Item(120, 30)  # Item 3: Value=120, Weight=30
]

knapsack_capacity = 50

print("--- Original Items ---")
for i, item in enumerate(items):
    print(f"Item {i+1}: {item}")

# Solve the knapsack problem
max_value, contents = fractional_knapsack(knapsack_capacity, items)

print(f"\n--- Knapsack Result ---")
print(f"Knapsack Capacity: {knapsack_capacity} kg")
print(f"Maximum Value Achieved: {max_value:.2f}")
print("Items in Knapsack:")
for content in contents:
    print(f"- {content}")
```

# Problem Definition
Given a set of
n
n items, each with:
Value vi
Weight wi
A knapsack with maximum capacity **W**

Goal: maximize total value that can be put in the knapsack.
Constraint: You may take fractions of items.

Greedy Strategy:

*  Sort items by decreasing ratio vi/wi
*  Select items fully until knapsack is full.
*  Take fraction of the next item if capacity remains.

In this lab, we implement Huffman Coding, a greedy algorithm used for lossless data compression.
It assigns variable-length binary codes to input symbols based on their frequencies, ensuring that more frequent symbols receive shorter codes and less frequent symbols receive longer codes.

This algorithm constructs a binary tree (Huffman Tree) where each leaf node represents a character, and the path from the root to a leaf gives the character’s binary code.

Use:
Huffman Tree Construction using a Min-Priority Queue (Min-Heap).
Code generation for each character.
Encoding and decoding using the generated Huffman codes.
Complexity analysis and comparison with fixed-length encoding.

# Pseudocode:
1. Create a leaf node for each character and insert all nodes into a
2. min-heap by frequency.
   *  While the heap has more than one node:
   * Extract two nodes with the lowest frequencies.
   * Create a new internal node with frequency = sum of the two.
   *  Set the two extracted nodes as left and right children.
   *  Insert the new node back into the min-heap.
3. The remaining node is the root of the Huffman Tree.

4. Traverse the tree:
   * Assign 0 for left edge and 1 for right edge.
   * The resulting paths from root to leaves give Huffman codes.

```python
import heapq
from collections import defaultdict

class Node:
    def __init__(self, char, freq, left=None, right=None):
        self.char = char
        self.freq = freq
        self.left = left
        self.right = right

    def __lt__(self, other):
        return self.freq < other.freq

def build_huffman_tree(text):
    if not text:
        return None

    freq = defaultdict(int)
    for char in text:
        freq[char] += 1

    min_heap = [Node(char, f) for char, f in freq.items()]
    heapq.heapify(min_heap)

    while len(min_heap) > 1:
        node1 = heapq.heappop(min_heap)
        node2 = heapq.heappop(min_heap)

        merged_node = Node(None, node1.freq + node2.freq, node1, node2)
        heapq.heappush(min_heap, merged_node)

    return min_heap[0] if min_heap else None

def generate_huffman_codes(node, current_code="", codes=None):
    if codes is None:
        codes = {}

    if node.char is not None:
        codes[node.char] = current_code if current_code else "0"
        return codes

    if node.left:
        generate_huffman_codes(node.left, current_code + "0", codes)
    if node.right:
        generate_huffman_codes(node.right, current_code + "1", codes)

    return codes

def huffman_encode(text, codes):
    encoded_text = ""
    for char in text:
        encoded_text += codes[char]
    return encoded_text

def huffman_decode(encoded_text, root_node):
    decoded_text = ""
    current_node = root_node

    for bit in encoded_text:
        if bit == '0':
            current_node = current_node.left
        else:
            current_node = current_node.right

        if current_node.char is not None:
            decoded_text += current_node.char
            current_node = root_node

    return decoded_text

text = "this is an example for huffman encoding"

root = build_huffman_tree(text)

huffman_codes = generate_huffman_codes(root)

print("Original Text:", text)
print("\nHuffman Codes:")
for char, code in sorted(huffman_codes.items()):
    print(f"  '{char}': {code}")

encoded_text = huffman_encode(text, huffman_codes)
print("\nEncoded Text:", encoded_text)
print(f"Encoded text length: {len(encoded_text)} bits")

decoded_text = huffman_decode(encoded_text, root)
print("\nDecoded Text:", decoded_text)

print(f"Original text matches decoded text: {text == decoded_text}")

fixed_length_bits = len(text) * 8
print(f"Original fixed-length text length: {fixed_length_bits} bits")
```
